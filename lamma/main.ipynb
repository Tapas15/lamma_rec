```python
import pandas as pd
```

```python
df = pd.read_csv('netflix_titles.csv')
```

```python
df
```

```python
def create_textual_representation(row):
    textual_representation = f"""Type: {row['type']},
    Title: {row['title']},
    Director: {row['director']},
    Cast: {row['cast']},
    Released: {row['release_year']},
    Genres: {row['listed_in']},
    Description: {row['description']}"""
    return textual_representation
```

```python
df['text_summary'] = df.apply(create_textual_representation, axis=1)
```

```python
print(df['text_summary'].values[0])
```

```python
import faiss
import requests
import numpy as np 
dim = 4096

# Correct FAISS index for L2 (Euclidean) distance
index = faiss.IndexFlatL2(dim)

# Initialize an empty numpy array of shape (num_samples, dim)
X = np.zeros((len(df['text_summary']), dim), dtype='float32')  # FAISS needs float32
```

```python
for i,representation in enumerate(df['text_summary']):
    if i % 200:
        print("processed",str(i),'instances')
    res = requests.post('http://localhost:11434/api/embeddings',
    json = {
                               'model':'llama3.2:latest',
                               'prompt':representation
                       }
    )


embedding = res.json()['embeddings']
X[i] = np.array(embedding)

index.add(X)

```

```python
index= faiss.write_index(index,'index')
```
